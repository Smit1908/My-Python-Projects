```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Create a flowchart for Power BI Project
fig, ax = plt.subplots(figsize=(10, 6))

# Remove axes
ax.axis("off")

# Define boxes (steps)
boxes = [
    ("Data Import\n(Excel/CSV)", (0.1, 0.7)),
    ("Data Cleaning\n(Power Query)", (0.40, 0.7)),
    ("Data Modeling\n& DAX", (0.7, 0.7)),
    ("Dashboard\n(Visuals & KPIs)", (0.35, 0.4)),
    ("Insights &\nDecision Making", (0.35, 0.1))
]

# Draw rectangles
for text, (x, y) in boxes:
    ax.add_patch(mpatches.FancyBboxPatch((x, y), 0.25, 0.15,
                                         boxstyle="round,pad=0.05",
                                         edgecolor="black", facecolor="#AED6F1"))
    ax.text(x + 0.125, y + 0.075, text, ha="center", va="center", fontsize=10, weight="bold")

# Draw arrows
arrowprops = dict(arrowstyle="->", lw=2, color="black")
ax.annotate("", xy=(0.35, 0.775), xytext=(0.1+0.25, 0.775), arrowprops=arrowprops)
ax.annotate("", xy=(0.6, 0.775), xytext=(0.35+0.25, 0.775), arrowprops=arrowprops)
ax.annotate("", xy=(0.475, 0.55), xytext=(0.475, 0.7), arrowprops=arrowprops)
ax.annotate("", xy=(0.475, 0.25), xytext=(0.475, 0.4), arrowprops=arrowprops)

plt.title("Power BI Sales Dashboard Project Flow", fontsize=14, weight="bold")
plt.show()
```
